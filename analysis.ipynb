#### 测试模块

1. 基本统计量：均值，胜率，稳定性(通过方差体现)
2. 分布估计和统计检验

```python
class ana:
    def __init__(self, res) -> None:
        self.data = res

    # basic
    def get_mean(self):
        return self.data.mean()
    
    def get_median(self):
        return self.data.median()

    def binarize(self):
        return self.data.apply(lambda x: 1 if x > 0 else 0)

    # statistical
    def draw_distribution(self):
        # colorful theme
        sns.set_theme(style="darkgrid")
        sns.distplot(self.data)
        plt.title('human+nfsp player VS sharp player')
        plt.show()
```

```python
ana(gb.get_group(('fake_player','nsfp_player'))['result']).draw_distribution()
```
